```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chisquare

p = [0.35, 0.15, 0.35, 0.15]
k = [3, 5, 7, 9]

n = 1000
sample = np.random.choice(k, size=n, p=p)

mean_empirical = np.mean(sample)
var_empirical = np.var(sample)

mean_theoretical = np.dot(p, k)
var_theoretical = np.dot(p, np.square(k)) - mean_theoretical**2

print(f"Эмпирическое математическое ожидание: {mean_empirical:.3f}")
print(f"Теоретическое математическое ожидание: {mean_theoretical:.3f}")
print(f"Эмпирическая дисперсия: {var_empirical:.3f}")
print(f"Теоретическая дисперсия: {var_theoretical:.3f}")

plt.hist(sample, bins=np.arange(2.5, 10, 1), density=True, rwidth=0.8, alpha=0.7, label="Гистограмма выборки")
plt.xticks(k)
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Генерация дискретной случайной величины')
plt.grid(True)
plt.legend()
plt.show()

sorted_sample = np.sort(sample)
y_vals = np.arange(1, len(sorted_sample)+1) / len(sorted_sample)

plt.step(sorted_sample, y_vals, label="Эмпирическая функция распределения", color='blue', where='post')
plt.scatter(k, np.cumsum(p), color='red', zorder=5, label="Теоретическая функция распределения")
plt.xlabel("Значение")
plt.ylabel("Функция распределения")
plt.legend()
plt.grid(True)
plt.title("Эмпирическая и теоретическая функции распределения")
plt.show()

observed_frequencies = np.array([np.sum(sample == ki) for ki in k])
expected_frequencies = np.array(p) * n
chi2_stat, p_value = chisquare(observed_frequencies, expected_frequencies)

print(f"Статистика χ²: {chi2_stat:.3f}")
print(f"p-значение: {p_value:.3f}")
```

```python
from IPython.display import display, Math
print("Теоретическое математическое ожидание:")
display(Math(r'\mathbb{E}[X] = \sum_{i=1}^{n} p_i \cdot k_i'))
print("Теоретическая дисперсия:")
display(Math(r'\text{Var}[X] = \sum_{i=1}^{n} p_i \cdot (k_i - \mathbb{E}[X])^2'))
print("Статистика хи-квадрат:")
display(Math(r'\chi^2 = \sum_{i=1}^{n} \frac{(f_{\text{observed},i} - f_{\text{expected},i})^2}{f_{\text{expected},i}}'))
print("p-значение:")
display(Math(r'p\text{-value} = 1 - F_{\chi^2}(\chi^2_{\text{stat}}, \text{df})'))
```
